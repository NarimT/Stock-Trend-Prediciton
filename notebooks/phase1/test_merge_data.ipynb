```python
nvda = nvda[nvda.index >= "2022-03-03"]
amd = amd[amd.index >= "2022-03-03"]
intc = intc[intc.index >= "2022-03-03"]
spy = spy[spy.index >= "2022-03-03"]
dia = dia[dia.index >= "2022-03-03"]
iwm = iwm[iwm.index >= "2022-03-03"]
btc = btc[btc.index >= "2022-03-03"]
oil = oil[oil.index >= "2022-03-03"]
gold = gold[gold.index >= "2022-03-03"]
nvda_insider = nvda_insider[nvda_insider["transaction_date"] >= "2022-03-03"]
```

```python
nvda_news["overall_sentiment_score"] = nvda_news["overall_sentiment_score"].astype(float)
nvda_news["nvda_sentiment_score"] = nvda_news["nvda_sentiment_score"].astype(float)
nvda_news["nvda_relevance_score"] = nvda_news["nvda_relevance_score"].astype(float)
```

```python
# combine nvda with news sentiment by aggregating news sentiment per hour, then merging on timestamp, left join
numeric_cols = nvda_news.select_dtypes(include="number").columns
nvda_news_agg = nvda_news[numeric_cols].resample("h").mean().reset_index()
nvda_news_agg = nvda_news_agg.rename(columns={"time_published": "timestamp"})
nvda_news_agg
```

```python
nvda_merged = pd.merge(nvda, nvda_news_agg, how="left", left_index=True, right_on="timestamp")
nvda_merged.index = nvda_merged["timestamp"]
nvda_merged = nvda_merged.drop(columns=["timestamp"])
nvda_merged
```

```python
# Shift index by +1 day after resampling, then forward fill
btc_hourly = btc.resample("h").mean()
btc_hourly.index = btc_hourly.index + pd.Timedelta(days=1)
btc_hourly = btc_hourly.ffill()

oil_hourly = oil.resample("h").mean()
oil_hourly.index = oil_hourly.index + pd.Timedelta(days=1)
oil_hourly = oil_hourly.ffill()

gold_hourly = gold.resample("h").mean()
gold_hourly.index = gold_hourly.index + pd.Timedelta(days=1)
gold_hourly = gold_hourly.ffill()
```

```python
# Reindex to nvda's timestamp index
btc_aligned = btc_hourly.reindex(nvda.index)
oil_aligned = oil_hourly.reindex(nvda.index)
gold_aligned = gold_hourly.reindex(nvda.index)

# Rename columns for clarity
btc_aligned = btc_aligned.rename(columns=lambda x: f"btc_{x.lower()}")
oil_aligned = oil_aligned.rename(columns=lambda x: f"oil_{x.lower()}")
gold_aligned = gold_aligned.rename(columns=lambda x: f"gold_{x.lower()}")

# Merge with nvda
nvda_merged_assets = pd.concat([nvda, btc_aligned, oil_aligned, gold_aligned], axis=1)
nvda_merged_assets
```

```python
# Map acquisition_or_disposal to +1 for 'A', -1 for 'D'
action_map = {"A": 1, "D": -1}
nvda_insider["action_sign"] = nvda_insider["acquisition_or_disposal"].map(action_map)

# Calculate signed shares
nvda_insider["signed_shares"] = nvda_insider["shares"] * nvda_insider["action_sign"]

# Group by transaction_date and calculate average signed shares per day
daily_avg_signed_shares = nvda_insider.groupby("transaction_date")["signed_shares"].mean()

daily_avg_signed_shares
```

```python
# Shift daily_avg_signed_shares index by +3 days
shifted_signed_shares = daily_avg_signed_shares.copy()
shifted_signed_shares.index = shifted_signed_shares.index + pd.Timedelta(days=3)

# Map to nvda index (hourly), forward fill within each day
nvda["signed_shares"] = nvda.index.normalize().map(shifted_signed_shares)
```
