```python
import yfinance as yf
```

```python
tickers = [
    "BTC-USD",  # Bitcoin USD
    "GC=F",  # Gold futures
    "CL=F",  # Crude Oil (WTI) futures
]
dfs = {}
for ticker in tickers:
    print(f"Downloading data for {ticker}...")
    df = yf.download(
        tickers=ticker,
        interval="1d",
        start="2020-01-01",
        end="2025-07-01",
        group_by="ticker",
        rounding=True,
        ignore_tz=False,
    )
    dfs[ticker] = df[ticker]
    print(f"Data for {ticker} downloaded, {len(df)} rows.")
```

```python
dfs["BTC-USD"].to_parquet("btc_usd_daily.parquet")
dfs["GC=F"].to_parquet("gold_futures_daily.parquet")
dfs["CL=F"].to_parquet("crude_oil_wti_futures_daily.parquet")
```
