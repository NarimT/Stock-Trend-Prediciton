```python
import joblib
```

```python
pipelines_1 = joblib.load("1_price_analysis/pipelines.joblib")
pipelines_2 = joblib.load("2_market_context_feature_analysis/pipelines.joblib")
pipelines_3 = joblib.load("3_other_feature_analysis/pipelines.joblib")
pipelines_4 = joblib.load("4_final_experiment/pipelines.joblib")
```

```python
import io
import joblib
import pandas as pd

rows = []
for key, pipeline in {**pipelines_1, **pipelines_2, **pipelines_3, **pipelines_4}.items():
    buffer = io.BytesIO()
    joblib.dump(pipeline, buffer)
    size_bytes = buffer.tell()
    size_kb = size_bytes / 1024
    window_size, feature_group, model_name = key
    rows.append(
        {"window_size": window_size, "feature_group": feature_group, "model_name": model_name, "size_kb": size_kb}
    )
df_sizes = pd.DataFrame(rows)
```

```python
df_sizes.groupby("model_name")["size_kb"].agg(["min", "max", "mean", "std"])
```
